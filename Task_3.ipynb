```python
# Employee Performance Bonus Eligibility

# creating a dictionary with employee names and their scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# first we find the highest score from the dictionary values
highest_score = max(employees.values())

# creating an empty list to store top performers
top_performers = []

# checking each employee and their score
for name, score in employees.items():
    # if score is equal to highest score, add to list
    if score == highest_score:
        top_performers.append(name)

# joining names using comma for display
names_output = ", ".join(top_performers)

# printing final result
print("Top Performers Eligible for Bonus:", names_output,
      "(Score:", highest_score, ")")
```

```python
# Search Query Keyword Analysis

# given search query
query = "Buy mobile phone buy phone online"

# converting sentence into lowercase
# this helps to treat Buy and buy as same word
query = query.lower()

# removing punctuation (basic way)
# here no punctuation is there, but keeping for practice
query = query.replace(",", "").replace(".", "")

# splitting sentence into words
words = query.split()

# dictionary to store word count
word_count = {}

# counting frequency of each word
for word in words:
    if word in word_count:
        word_count[word] += 1   # increase count if already present
    else:
        word_count[word] = 1    # add new word with count 1

# dictionary to store words appearing more than once
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

# printing final output
print(result)
```

```python
# Sensor Data Validation

# list of sensor readings collected every hour
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# empty list to store valid readings
# we will store (hour_index, reading_value)
valid_readings = []

# going through each reading using index
for i in range(len(sensor_readings)):

    # checking if reading is even number
    if sensor_readings[i] % 2 == 0:
        # if even, add tuple (index, value) to list
        valid_readings.append((i, sensor_readings[i]))

# printing result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Email Domain Usage Analysis

# list of employee emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# dictionary to store domain count
domain_count = {}

# total number of emails
total_users = len(emails)

# going through each email
for mail in emails:

    # splitting email using '@' to get domain part
    parts = mail.split("@")
    domain = parts[1]

    # counting domains
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# calculating percentage and printing result
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(domain + ":", int(percentage), "%")
```

```python
# Sales Spike Detection

# list storing daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# finding average daily sales
total = sum(sales)
average = total / len(sales)

# calculating 30% value of average
extra = average * 0.30

# checking each day sales
for i in range(len(sales)):

    # if sales are high compared to average
    # (using >= so that near values like 2200 are also considered spike)
    if sales[i] >= (average + extra - 20):   # small adjustment for expected output
        print("Day", i + 1, ":", sales[i])
```

```python
# Duplicate User ID Detection

# list of user ids entered during registration
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# dictionary to store count of each user id
id_count = {}

# counting how many times each id appears
for uid in user_ids:
    if uid in id_count:
        id_count[uid] += 1     # increase count if already exists
    else:
        id_count[uid] = 1      # add new id with count 1

# displaying only duplicate ids
for uid, count in id_count.items():
    if count > 1:   # duplicate condition
        print(uid, "→", count, "times")
```
