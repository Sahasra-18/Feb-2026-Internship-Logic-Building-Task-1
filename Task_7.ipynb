```python
# Smart Light Controller Program

# Creating a class for Smart Light
class SmartLight:

    # constructor to initialize light name and status
    def __init__(self, name):
        self.name = name      # store light name
        self.status = "OFF"   # light is OFF by default

    # function to turn ON the light
    def turn_on(self):
        self.status = "ON"

    # function to turn OFF the light
    def turn_off(self):
        self.status = "OFF"

    # function to display current status
    def show_status(self):
        print(self.name + " is " + self.status + ".")


# ---- Main Program ----

# taking input from user
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

# creating object of SmartLight class
light = SmartLight(light_name)

# checking action and performing operation
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()
else:
    print("Invalid Action")

# displaying final status
light.show_status()
```

```python
# Employee ID Card System

# creating a class for Employee
class Employee:

    # constructor to store employee details
    def __init__(self, name, emp_id, department):
        self.name = name              # employee name
        self.emp_id = emp_id          # employee ID
        self.department = department  # department name

    # function to display ID card details
    def display_id_card(self):
        print("\n----- EMPLOYEE ID CARD -----")
        print("Name:", self.name)
        print("Employee ID:", self.emp_id)
        print("Department:", self.department)


# ---- Main Program ----

# taking input from user
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# creating object
emp = Employee(name, emp_id, department)

# displaying details
emp.display_id_card()
```

```python
# Mobile Contact Record Program

# creating a class for Contact
class Contact:

    # constructor to store contact details
    def __init__(self, name, phone):
        self.name = name      # store contact name
        self.phone = phone    # store phone number

    # function to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# ---- Main Program ----

# taking input from user
name = input("Contact Name: ")
phone = input("Phone Number: ")

# creating object of Contact class
c1 = Contact(name, phone)

# displaying saved contact
c1.display_contact()
```

```python
# Product Price Tag Generator

# creating a class for Product
class Product:

    # constructor to store product details
    def __init__(self, name, price):
        self.name = name      # product name
        self.price = price    # product price

    # function to print price tag
    def show_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# ---- Main Program ----

# taking input from user
product_name = input("Product Name: ")
price = input("Price: ")

# creating object of Product class
p1 = Product(product_name, price)

# displaying price tag
p1.show_price_tag()
```

```python
# Movie Rating Display System

# creating a class for Movie
class Movie:

    # constructor to store movie details
    def __init__(self, name, rating):
        self.name = name        # movie name
        self.rating = rating    # movie rating

    # function to display movie details
    def show_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# ---- Main Program ----

# taking input from user
movie_name = input("Movie Name: ")
rating = input("Rating: ")

# creating object of Movie class
m1 = Movie(movie_name, rating)

# displaying movie information
m1.show_details()
```

```python
# Delivery Address Manager

# creating a class for Delivery details
class Delivery:

    # constructor to store customer details
    def __init__(self, name, address):
        self.name = name        # customer name
        self.address = address  # delivery address

    # function to display delivery details
    def show_delivery(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)


# ---- Main Program ----

# taking input from user
customer_name = input("Customer Name: ")
address = input("Address: ")

# creating object of Delivery class
d1 = Delivery(customer_name, address)

# displaying delivery information
d1.show_delivery()
```
